```python
from scipy.io import arff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rng # random number generator
import random
from sklearn.model_selection import train_test_split
import numpy.random as rng # random number generator
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error
```

```python
import numpy.random as rng # random number generator

print(rng.random())
#random_state simply sets a seed to the random generato
rn = rng.RandomState(47)
print(rn.random())
```

```python
#Steel plates Data 
steel_plates_fault = arff.loadarff('steel-plates-fault.arff')
steel = pd.DataFrame(steel_plates_fault[0])
steel.Class = steel.Class.str.decode('utf-8')
steel_y = steel.Class
steel_x = steel.drop('Class', axis=1)
steel.head()
```

```python
# Ionoshpere Data
ionoshpere_dic = arff.loadarff('dataset_59_ionosphere.arff')
ionosphere = pd.DataFrame(ionoshpere_dic[0])
ionosphere['class'] = ionosphere['class'].str.decode('utf-8')
iono_y = ionosphere['class']
iono_x = ionosphere.drop(['class'], axis=1)
ionosphere.head()
```

```python
banknotes = arff.loadarff('banknotes.arff')
bank = pd.DataFrame(banknotes[0])
bank.Class = bank.Class.str.decode('utf-8')
bank_y = bank.Class
bank_x = bank.drop('Class', axis=1)
bank.head()
```

```python
###########################################################################
#######  K Nearest Neighbours Classifier for the Steel plates Data ########
###########################################################################
from sklearn.neighbors import KNeighborsClassifier
rn = rng.RandomState(47)


neigh1= KNeighborsClassifier(n_neighbors=1)
neigh2= KNeighborsClassifier(n_neighbors=2)
neigh3= KNeighborsClassifier(n_neighbors=3)
neigh4= KNeighborsClassifier(n_neighbors=4)
neigh5= KNeighborsClassifier(n_neighbors=5)


#############################################
#Steel Plates 
#Steel plates Fault KNN=1
neigh1steel = []
for i in range(50):
    strain_x, stest_x, strain_y, stest_y = train_test_split(steel_x, steel_y, test_size=0.5, random_state=rn)
    neigh1.fit(strain_x, strain_y)
    spre_y1 = neigh1.predict(stest_x)
    neigh1steel = np.append(neigh1steel, accuracy_score(stest_y, spre_y1))
    
#Steel plates Fault KNN=2
neigh2steel = []
for i in range(50):
    strain_x, stest_x, strain_y, stest_y = train_test_split(steel_x, steel_y, test_size=0.5, random_state=rn)
    neigh2.fit(strain_x, strain_y)
    spre_y2 = neigh2.predict(stest_x)
    neigh2steel = np.append(neigh2steel, accuracy_score(stest_y, spre_y2))
    
#Steel plates Fault KNN=3
neigh3steel = []
for i in range(50):
    strain_x, stest_x, strain_y, stest_y = train_test_split(steel_x, steel_y, test_size=0.5, random_state=rn)
    neigh3.fit(strain_x, strain_y)
    spre_y3 = neigh3.predict(stest_x)
    neigh3steel = np.append(neigh3steel, accuracy_score(stest_y, spre_y3))
    
#Steel plates Fault KNN=4
neigh4steel = []
for i in range(50):
    strain_x, stest_x, strain_y, stest_y = train_test_split(steel_x, steel_y, test_size=0.5, random_state=rn)
    neigh4.fit(strain_x, strain_y)
    spre_y4 = neigh4.predict(stest_x)
    neigh4steel = np.append(neigh4steel, accuracy_score(stest_y, spre_y4))

#Steel plates Fault KNN=5
neigh5steel = []
for i in range(50):
    strain_x, stest_x, strain_y, stest_y = train_test_split(steel_x, steel_y, test_size=0.5, random_state=rn)
    neigh5.fit(strain_x, strain_y)
    spre_y5 = neigh5.predict(stest_x)
    neigh5steel = np.append(neigh5steel, accuracy_score(stest_y, spre_y5))  
    

#producing the boxplot
steelKNNs = (neigh1steel, neigh2steel, neigh3steel, neigh4steel, neigh5steel)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(steelKNNs)

ax.set_title('KNN on steel fault data')
ax.set_xlabel('Hyperparameter (n_neighbors)')
ax.set_ylabel('Accuracy')

plt.show()

knn_mse_steel = [1-neigh1steel,1-neigh2steel,1-neigh3steel,1-neigh4steel,1-neigh5steel]
knn_mse_steel= pd.DataFrame(knn_mse_steel)
knn_mse_steel.min().min()
```

```python
import sklearn.datasets as dt
from sklearn.cluster import KMeans
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.cluster import Birch
```

```python
#creating the datasets 
blobs_X, blobs_Y = dt.make_blobs(n_samples=1000, n_features=2)
class_X, class_y = dt.make_classification(n_samples=1000, n_clusters_per_class=1, n_features=2,n_redundant=0)
circles_X, circles_Y = dt.make_circles(n_samples=1000, noise = 0.3)
```

```python
# K means clustering
rn = rng.RandomState(47)

kmeans_blobs = KMeans(n_clusters= 3, random_state=rn)
kmeans = KMeans(n_clusters= 2, random_state=rn)

km_blobs = kmeans_blobs.fit(blobs_X)
km_class = kmeans.fit(class_X)
km_circles = kmeans.fit(circles_X)
```

```python
#K means clustering for Blobs
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(blobs_X[:,0], blobs_X[:,1], c=km_blobs.labels_,cmap=plt.cm.prism)
ax.set_title('$k$-Means Clustering on Blobs, K=3');
```

```python
#K means clustering for Class
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(class_X[:,0], class_X[:,1], c=km_class.labels_, cmap=plt.cm.prism)
ax.set_title('$k$-Means Clustering on class, K=2');
```

```python
#K means clustering for Cricles
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(circles_X[:,0], circles_X[:,1], c=km_circles.labels_, cmap=plt.cm.prism)
ax.set_title('$k$-Means Clustering on Circles, K=2');
```

```python
#Affinity Propagation Clustering 
rn = rng.RandomState(47)

af_clustering = AffinityPropagation(random_state=rn)

af_blobs = af_clustering.fit(blobs_X)
af_class = af_clustering.fit(class_X)
af_circles = af_clustering.fit(circles_X)
```
